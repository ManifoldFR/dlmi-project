```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

```python
files = glob.glob("*.csv")
```

```python
files
```

```python
dfs = [pd.read_csv(f) for f in files]
```

```python
fig = plt.figure()
plt.plot(dfs[0].Value, label="Full RGB")
plt.plot(dfs[1].Value, label="Green channel only")
plt.title("Validation Dice score")
plt.ylim((0.4, 0.85))
plt.legend()
plt.tight_layout()
```

```python
fig.savefig('unet-rgb-green-comparison.pdf')
```
